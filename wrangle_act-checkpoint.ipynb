## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html).
```python
# Make copies of original pieces of data
df1_clean = df1.copy()
df2_clean = df2.copy()
df3_clean = df3.copy()
```

###  Tidiness Issue #1:

As adviced by the Udacity Instructor, let's start with Tidiness Issue #1

#### Define:

The 4 dog stages column (doggo, floofer, pupper, puppo) needs to be under one column 'stage' with four categories

#### Code

```python
#Let's replace none and empty strings in the following stages column
df1_clean.doggo.replace('None', '', inplace = True)
df1_clean.floofer.replace('None', '', inplace = True)                       
df1_clean.pupper.replace('None', '', inplace = True)
df1_clean.puppo.replace('None', '', inplace = True)

#Combining stages column into a new colum 'stage'
df1_clean['stage'] = df1_clean.doggo + df1_clean.floofer + df1_clean.pupper + df1_clean.puppo

#Using loc function to format the entries of combination of 2 dog stages
df1_clean.loc[df1_clean.stage == 'doggopupper', 'stage'] = 'doggo, pupper'
df1_clean.loc[df1_clean.stage == 'doggopuppo', 'stage'] = 'doggo, puppo'
df1_clean.loc[df1_clean.stage == 'doggofloofer', 'stage'] = 'doggo, flooper'

#Dropping the old columns
df1_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1, inplace = True)

#Filling the empty spaces
df1_clean['stage'] = df1_clean['stage'].replace('', np.nan)

```
